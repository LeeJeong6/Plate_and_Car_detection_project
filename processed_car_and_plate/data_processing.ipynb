car랑 plate랑 이미지가 몇개가 일치할까? ->285개 , car_image_paths, label_paths

```python
import re
import os

total = 0

# 자동차 폴더와 번호판 폴더 경로 설정
car_folder = "/mnt/hdd_6tb/seungeun/HuNature/data/image/car"
license_folder = "/mnt/hdd_6tb/seungeun/HuNature/data/image/plate"

# 정규 표현식 패턴 정의
car_pattern = re.compile(r"C-\d+_(\d+)_SR(\d{2})_01_A(\d{4})\.jpg")
license_pattern = re.compile(r"C-\d+_(\d+)_SR(\d{2})_01_N(\d{4})\.jpg")

# 모든 하위 폴더의 파일 탐색 함수
def get_all_files(folder_path):
    file_paths = []
    for root, _, files in os.walk(folder_path):
        for file in files:
            if file.endswith(".jpg"):
                file_paths.append(os.path.join(root, file))
    return file_paths

# 자동차와 번호판 파일 매칭
def match_files(car_folder, license_folder):
    car_files = get_all_files(car_folder)
    license_files = get_all_files(license_folder)

    matches = []

    # 파일 정보를 딕셔너리에 저장
    car_dict = {}
    license_dict = {}

    for car_file in car_files:
        match = car_pattern.search(os.path.basename(car_file))
        if match:
            key = (match.group(1), match.group(2), match.group(3))
            car_dict[key] = car_file

    for license_file in license_files:
        match = license_pattern.search(os.path.basename(license_file))
        if match:
            key = (match.group(1), match.group(2), match.group(3))
            license_dict[key] = license_file

    # 자동차 파일과 번호판 파일 매칭 확인
    for key in car_dict:
        if key in license_dict:
            matches.append((car_dict[key], license_dict[key]))

    return matches

# 매칭 실행
matched_files = match_files(car_folder, license_folder)

# 이미지 경로 리스트와 라벨 경로 리스트 생성
car_image_paths = []
label_paths = []
plate_label_paths=[]
for car_file, license_file in matched_files:
    total += 1
    # 자동차 이미지 경로 저장
    car_image_paths.append(car_file)
    # VS -> VL 치환 및 확장자 json 변경 후 저장
    label_path = car_file.replace('/VS_', '/VL_').replace('/image/', '/label/').rsplit('.', 1)[0] + '.json'
    label_paths.append(label_path)
    plate_label_path = license_file.replace('/VS_', '/VL_').replace('/image/', '/label/').rsplit('.', 1)[0] + '.json'
    plate_label_paths.append(plate_label_path)
print(f"총 {total}개의 매칭된 파일:")
print("자동차 이미지 경로 리스트:")
print(plate_label_paths)
```

3.이미지 crop하기

```python
from PIL import Image
import cv2
import natsort
import os

image_dir = "/mnt/hdd_6tb/bill0914/processed_car_and_plate/matched_car"
label_dir = "/mnt/hdd_6tb/bill0914/processed_car_and_plate/matched_label"
out_dir = "/mnt/hdd_6tb/bill0914/processed_car_and_plate/matched_carcropimg"

image_list = natsort.natsorted(os.listdir(image_dir))
label_list = natsort.natsorted(os.listdir(label_dir))

for i, image_name in enumerate(image_list):
    image_path = os.path.join(image_dir, image_name)
    image = cv2.imread(image_path)  # 이미지 불러오기 (array 형태)
    
    if image is None:
        print(f"이미지를 불러올 수 없습니다: {image_path}")
        continue
    
    height, width, _ = image.shape  # 이미지의 높이와 너비 가져오기
    
    # 해당 이미지의 라벨 읽기
    label_path = os.path.join(label_dir, label_list[i])
    with open(label_path, 'r') as f:
        carbbox_list = f.readlines()
    
    for idx, g in enumerate(carbbox_list):
        # 라벨의 좌상단 x, y와 w, h 파싱
        class_id, x, y, w, h = map(float, g.split())
        
        # 좌표를 정수로 변환
        x1 = int(x)
        y1 = int(y)
        x2 = int(x + w)
        y2 = int(y + h)
        
        # 이미지 크롭 (좌표가 이미지 경계를 벗어나지 않도록 클램핑)
        x1 = max(0, x1)
        y1 = max(0, y1)
        x2 = min(width, x2)
        y2 = min(height, y2)
        
        car_image = image[y1:y2, x1:x2]  # 크롭된 이미지
        
        if car_image.size == 0:
            print(f"크롭된 이미지가 비어 있습니다: {image_name}, 라벨 {idx}")
            continue
        
        # 크롭된 이미지를 PIL 형식으로 변환 후 리사이즈
        car_image = Image.fromarray(cv2.cvtColor(car_image, cv2.COLOR_BGR2RGB))
        
        
        # 파일 이름 생성 및 저장
        name = f"car_image_{i + 1}_{idx}.png"
        car_image.save(os.path.join(out_dir, name))
```
