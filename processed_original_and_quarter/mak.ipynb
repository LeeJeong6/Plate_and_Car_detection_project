1. 주어진 데이터셋 이름과 크기 확인하기

```python
import os
import natsort
from PIL import Image
image_root = "/mnt/hdd_6tb/seungeun/HuNature/data/image/plate"
label_root = "/mnt/hdd_6tb/seungeun/HuNature/data/label/plate"
sorted_image_folder_list = natsort.natsorted(os.listdir(image_root))
sorted_label_folder_list = natsort.natsorted(os.listdir(label_root))

for idx,subfolder in enumerate(sorted_image_folder_list):
    image_path = os.path.join(image_root,subfolder)
    image_list = os.listdir(image_path)
    first_image_file_root = os.path.join(image_path,image_list[0])
    name,ext = os.path.splitext(first_image_file_root)
    image_size = Image.open(first_image_file_root).size
    print(f"Image데이터 {idx+1}번째 폴더의 이름: {os.path.basename(image_path)} 데이터 형태 : {ext}\n데이터 개수는 : {len(image_list)}, 이미지크기: {image_size}")
print("\n")
for idx,subfolder in enumerate(sorted_label_folder_list):
    label_path = os.path.join(label_root,subfolder)
    label_list = os.listdir(label_path) 
    first_label_file_root = os.path.join(label_path,label_list[0])
    name,ext = os.path.splitext(first_label_file_root)
    print(f"Label데이터 {idx+1}번째 폴더의 이름: {os.path.basename(label_path)} 데이터 형태 : {ext}\n데이터 개수는 : {len(label_list)}")

```

2.label데이터에서 BBOX좌표 추출하기

```python
import json
import os
from PIL import Image
image_root = "/mnt/hdd_6tb/seungeun/HuNature/data/image/plate/VS_차량번호판인식_교차로_[cr01]비산사거리_04번"
image_list = os.listdir(image_root)
image_list = natsort.natsorted(image_list)
first_image_path = os.path.join(image_root,image_list[77])
img = Image.open(first_image_path)
result_path = os.path.join("/mnt/hdd_6tb/bill0914/result","result.png")
img.save(result_path)


bbox_list=[]
labels_root = "/mnt/hdd_6tb/seungeun/HuNature/data/label/plate/VL_차량번호판인식_교차로_[cr01]비산사거리_04번"
labels_list = os.listdir(labels_root)
labels_list = natsort.natsorted(labels_list)
first_label_path = os.path.join(labels_root,labels_list[77])
with open(first_label_path) as file:
    datas = json.load(file)
    for annotations in datas["Learning_Data_Info"]["annotations"] :
        for a in annotations.get("license_plate"):
            bbox = a.get("bbox")
            x,y,h,w = bbox
            bbox_list.append(bbox)

txt_path = "/mnt/hdd_6tb/bill0914/result/result.txt"
with open(txt_path,mode='w') as f:
    for idx,bbox in enumerate(bbox_list):     
        if idx<len(bbox_list)-1:
            f.write(f"{0} {' '.join(map(str,bbox))}\n")
        else:
            f.write(f"{0} {' '.join(map(str,bbox))}")    


```

테스트데이터 라벨 다운받아서 저장하기

```python
import os
import json
import natsort
import json

file=1
car_label_dir = "/mnt/hdd_6tb/seungeun/HuNature/Test/02.라벨링데이터" #이거 바꾸자~~~~~~~~~~~~~~
save_dir = "/mnt/hdd_6tb/bill0914/mak/test_label_bbox" #이것두
for subfolder in natsort.natsorted(os.listdir(car_label_dir)):
    subfolder_path = os.path.join(car_label_dir,subfolder) #car/VL~~~

    for jsons in natsort.natsorted(os.listdir(subfolder_path)):
        jsons_path = os.path.join(subfolder_path,jsons)

        for i in natsort.natsorted(os.listdir(jsons_path)):
            jp = os.path.join(jsons_path,i)

            with open(jp,'r') as f:
                data = json.load(f)
                data = data['Learning_Data_Info']['annotations']
                print(len(data[0].get('license_plate'))) #bbox개수
                print(data[0].get('license_plate')[0].get('bbox'))

                for i in range(len(data[0].get('license_plate'))) :
                    bbox_list = data[0].get('license_plate')[i].get('bbox')
                    x = bbox_list[0]
                    y = bbox_list[1]
                    w = bbox_list[2]
                    h = bbox_list[3]
                    print(x,y,w,h)
                    file_name = f"img_{file}.txt"
                    save_path = os.path.join(save_dir,file_name)
                    with open(save_path,'a') as f:
                        if i ==len(data[0].get('license_plate'))-1 : 
                            f.write(f"{x} {y} {w} {h}")
                        else:
                            f.write(f"{x} {y} {w} {h}\n") 


            file+=1    
```
