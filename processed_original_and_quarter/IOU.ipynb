inference - 예측한 결과의 bbox좌표를 저장 -mak폴더

```python
from ultralytics import YOLO
import os
import natsort

# 학습된 모델 불러오기
model = YOLO('/mnt/hdd_6tb/bill0914/processed_original/runs/detect/train2/weights/best.pt')

# 테스트할 이미지 폴더 경로
save_folder = "/mnt/hdd_6tb/bill0914/processed_original/test"
input_folder = "/mnt/hdd_6tb/seungeun/HuNature/Test/01.원천데이터"
output_dir = "/mnt/hdd_6tb/bill0914/mak/예측한 결과의 bbox좌표"
input_folder_list = natsort.natsorted(os.listdir(input_folder))  # 5개 [cr06][cr11][cr12][cr13][cr14]
idx = 1

for subfolder in input_folder_list:
    subfolder_path = os.path.join(input_folder, subfolder)  # /mnt/hdd_6tb/seungeun/HuNature/Test/01.원천데이터/cr06
    subfolder_list = natsort.natsorted(os.listdir(subfolder_path))  # 01번 02번 03번 04번
    for jpgs in subfolder_list:
        jpgs_path = os.path.join(subfolder_path, jpgs)
        for js in natsort.natsorted(os.listdir(jpgs_path)):
            results = model(os.path.join(jpgs_path, js))
            output_file = os.path.join(output_dir, f"img_{idx}.txt")

            # 파일을 한 번만 열고, 그 안에 여러 줄을 쓰도록 변경
            with open(output_file, 'w') as f:  # 한 번만 열어서 처리
                for result in results:
                    boxes = result.boxes  # Boxes object for bounding box outputs
                    for box in boxes:  # Iterate through each box
                        xyxy = box.xyxy[0].cpu().numpy()  # Extract xyxy coordinates and move to CPU
                        # Write each bounding box in a new line for each image
                        f.write(f"{xyxy[0]} {xyxy[1]} {xyxy[2]} {xyxy[3]}\n")

            idx += 1
```

```python
ld="/mnt/hdd_6tb/bill0914/processed_original/test/label_bbox_count"
td="/mnt/hdd_6tb/bill0914/processed_quarter/test/test_count"
import os
from natsort import natsorted
ld_list = natsorted(os.listdir(ld))
td_list = natsorted(os.listdir(td))
ld_win=0
ld_listt=[]
ld_win_list=[]
same=0
same_list=[]
td_win=0
td_win_list=[]
td_listt=[]
num=0
for i in range(len(ld_list)):
    with open(os.path.join(ld,ld_list[i]),'r') as f:
        ld_count = f.read()
        ld_count = int(ld_count)
    with open(os.path.join(td,td_list[i]),'r') as f:
        td_count = f.read()
        td_count = int(td_count)
    
    if ld_count <td_count:
        td_listt.append(td_list[i].split('_')[1])
        td_win+=1 #데이터 개수(총485개)
        td_win_list.append(td_count-ld_count) #이긴 리스트  
        
       
    if ld_count >td_count:   
        ld_listt.append(ld_list[i].split('_')[2])
        ld_win+=1
        ld_win_list.append(ld_count-td_count)
        

    if ld_count == td_count:
        same+=1
        same_list.append(td_list[i].split('_')[1])
        num+=td_count

print(num)

print(td_listt)    
print(td_win)    
print(td_win_list)

print(ld_listt)
print(ld_win)
print(ld_win_list)

print(same_list)
print(same)

```
