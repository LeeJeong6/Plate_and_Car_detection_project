```python
import os
import shutil

# 원본 이미지와 라벨 폴더
train_image_dir = "/mnt/hdd_6tb/bill0914/processed_original/data/train/image"
val_image_dir = "/mnt/hdd_6tb/bill0914/processed_original/data/val/image"
label_dir = "/mnt/hdd_6tb/bill0914/processed_original/label"

# 복사할 목표 폴더
train_label_dir = "/mnt/hdd_6tb/bill0914/processed_original/data/train/label"
val_label_dir = "/mnt/hdd_6tb/bill0914/processed_original/data/val/label"

# 디렉토리 생성 (없으면)
os.makedirs(train_label_dir, exist_ok=True)
os.makedirs(val_label_dir, exist_ok=True)

# 이미지 번호 추출 함수
def extract_image_number(image_path):
    # 예시: image1.png에서 숫자 1을 추출
    return os.path.splitext(os.path.basename(image_path))[0].replace('image', '')

# 이미지 폴더에서 이미지 파일을 가져오기
train_image_files = [f for f in os.listdir(train_image_dir) if f.endswith('.png')]
val_image_files = [f for f in os.listdir(val_image_dir) if f.endswith('.png')]

# 이미지 번호 기준으로 라벨 파일 복사
def copy_labels(image_files, image_dir, label_dir, output_label_dir):
    for image_file in image_files:
        image_number = extract_image_number(image_file)
        label_file = f"image{image_number}.txt"
        label_path = os.path.join(label_dir, label_file)
        if os.path.exists(label_path):  # 라벨 파일이 존재하는 경우
            output_label_path = os.path.join(output_label_dir, label_file)
            shutil.copy(label_path, output_label_path)
            print(f"라벨 파일 {label_file} 복사 완료!")
        else:
            print(f"라벨 파일 {label_file}가 존재하지 않습니다.")

# train 이미지에 해당하는 라벨 복사
copy_labels(train_image_files, train_image_dir, label_dir, train_label_dir)

# val 이미지에 해당하는 라벨 복사
copy_labels(val_image_files, val_image_dir, label_dir, val_label_dir)

print("라벨 파일 복사 완료!")
```

각 폴더에 몇개의 데이터가 있는지

```python
import os

def count_files_in_directory(path, extensions=None):
    
    if not os.path.exists(path):
        raise ValueError(f"경로가 존재하지 않습니다: {path}")
    if not os.path.isdir(path):
        raise ValueError(f"입력 경로가 디렉토리가 아닙니다: {path}")
    
    # 디렉토리 내 파일 필터링 및 개수 계산
    file_count = 0
    for file in os.listdir(path):
        file_path = os.path.join(path, file)
        if os.path.isfile(file_path): 
            if extensions is None or file.lower().endswith(extensions):
                file_count += 1
    
    return file_count


path = "/mnt/hdd_6tb/seungeun/HuNature/Test/01.원천데이터"  # 경로 입력
extensions = (".txt", ".png")  # 원하는 확장자
file_count = count_files_in_directory(path, extensions)

print(f"경로 '{path}'에 있는 파일 개수: {file_count}")
```

빠진게 뭐인가 하는 

```python
import os

def find_missing_files(train_path, val_path, original_path):
    """
    train, val 디렉토리의 파일들을 합쳐서 original 디렉토리와 비교하여 누락된 파일을 찾습니다.
    
    Args:
        train_path (str): train 디렉토리 경로.
        val_path (str): val 디렉토리 경로.
        original_path (str): 원본 label 디렉토리 경로.
        
    Returns:
        list: 누락된 파일 목록.
    """
    def get_file_set(path):
        """디렉토리 내 파일 이름(확장자 제외) 집합 반환."""
        return {os.path.splitext(file)[0] for file in os.listdir(path) if os.path.isfile(os.path.join(path, file))}
    
    # 각 디렉토리에서 파일 목록 가져오기
    train_files = get_file_set(train_path)
    val_files = get_file_set(val_path)
    original_files = get_file_set(original_path)
    
    # train과 val의 파일을 합친 전체 파일 집합
    combined_files = train_files.union(val_files)
    
    # 원본 파일 집합과 비교하여 누락된 파일 찾기
    missing_files = original_files - combined_files
    return missing_files

# 경로 설정
train_label_path = "/mnt/hdd_6tb/bill0914/processed_original/data/train/label"
val_label_path = "/mnt/hdd_6tb/bill0914/processed_original/data/val/label"
original_label_path = "/mnt/hdd_6tb/bill0914/processed_original/label"

# 누락된 파일 확인
missing_files = find_missing_files(train_label_path, val_label_path, original_label_path)

if missing_files:
    print("누락된 파일 목록:")
    for file in sorted(missing_files):
        print(file)
else:
    print("모든 파일이 train과 val에 존재합니다.")
```

val만들어서 split하기 이때 파일이 없다면 없다고 프린팅


```python
import os
import shutil
import random

# 원본 디렉토리
label_dir = "/mnt/hdd_6tb/bill0914/processed_original/label"
image_dir = "/mnt/hdd_6tb/bill0914/processed_original/image"

# 복사할 목표 디렉토리
train_label_dir = "/mnt/hdd_6tb/bill0914/processed_original/data/train/labels"
val_label_dir = "/mnt/hdd_6tb/bill0914/processed_original/data/val/labels"
train_image_dir = "/mnt/hdd_6tb/bill0914/processed_original/data/train/images"
val_image_dir = "/mnt/hdd_6tb/bill0914/processed_original/data/val/images"

# 디렉토리 생성 (없으면 생성)
os.makedirs(train_label_dir, exist_ok=True)
os.makedirs(val_label_dir, exist_ok=True)
os.makedirs(train_image_dir, exist_ok=True)
os.makedirs(val_image_dir, exist_ok=True)

# 라벨 파일 목록 가져오기
label_files = [f for f in os.listdir(label_dir) if f.endswith('.txt')]

# 라벨 파일 섞기 (랜덤 셔플)
random.shuffle(label_files)

# 75%는 train, 25%는 val로 분할
split_index = int(len(label_files) * 0.75)
train_files = label_files[:split_index]
val_files = label_files[split_index:]

# 복사 함수
def copy_files(file_list, src_label_dir, src_image_dir, dest_label_dir, dest_image_dir):
    for label_file in file_list:
        # 라벨 파일 경로
        src_label_path = os.path.join(src_label_dir, label_file)
        dest_label_path = os.path.join(dest_label_dir, label_file)
        
        # 이미지 파일 경로
        image_file = label_file.replace('.txt', '.png')  # 라벨 파일명을 이미지 파일명으로 변환
        src_image_path = os.path.join(src_image_dir, image_file)
        dest_image_path = os.path.join(dest_image_dir, image_file)
        
        # 라벨 복사
        shutil.copy(src_label_path, dest_label_path)
        
        # 이미지가 존재하면 복사
        if os.path.exists(src_image_path):
            shutil.copy(src_image_path, dest_image_path)
        else:
            print(f"이미지 파일 {image_file}가 존재하지 않습니다.")

# Train 데이터 복사
copy_files(train_files, label_dir, image_dir, train_label_dir, train_image_dir)

# Val 데이터 복사
copy_files(val_files, label_dir, image_dir, val_label_dir, val_image_dir)

print(f"Train 라벨 개수: {len(train_files)}")
print(f"Val 라벨 개수: {len(val_files)}")
```
